```python
def corr(cov):
    corr = np.zeros(cov.shape)
    for i in range(cov.shape[0]):
        for j in range(cov.shape[1]):
            corr[i,j] = cov[i,j]/(np.sqrt(cov[i,i]*cov[j,j]))
    return corr
```
