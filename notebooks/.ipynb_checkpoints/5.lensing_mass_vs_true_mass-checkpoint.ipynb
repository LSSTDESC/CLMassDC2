```python
%load_ext autoreload
%autoreload 2
import CL_WL_fit_cluster_mass_v2 as fit
import CL_Mass_richness_relation as mr
MR_modeling = mr.WL_Mass_Richness()
MR_modeling.set_pivot_values(analysis.z0, analysis.richness0)
```

```python
sigma_lnm = .15
scatter_lnc = .2
logmass_individual = []
radius = profile_sky_stack[0]['radius']
for i, p in enumerate(profile_sky_stack):
    p_ = []
    richness_individual = p['richness']
    redshift_individual = p['redshift']
    n = len(redshift_individual)
    logm = np.random.randn(len(richness_individual))*sigma_lnm/(np.log(10)) + MR_modeling.mu_logM_lambda(richness_individual, redshift_individual, [14.1, -.17, .7])
    for j, r in enumerate(richness_individual):
        c = fit.concDiemer15._concentration(cosmo_ccl, 10**logm[j], 1./(1. + redshift_individual[j]))
        c = np.exp(np.log(c) + scatter_lnc*np.random.random())
        gt = clmm.compute_excess_surface_density(radius, 10**logm[j], c, redshift_individual[j], cosmo)
        p_.append(gt + np.random.random(len(gt))*covariance_sky_stack[i]['cov_t'].diagonal()**.5)
    
    stack_true = np.average(np.array(p_), weights = p['W_l_true'], axis=0)
    plt.loglog(stack_true)
```

```python
def esd_modeling(R, logm, c, z, cosmo):
    return clmm.compute_excess_surface_density(R, 10**logm, c, z, cosmo, delta_mdef=200,
                                       halo_profile_model='nfw', massdef='critical')
```

```python
lnL.reshape_data(r_min = 1, r_max = 5.5)
```

```python
def lnLikelihood(p):
    logm0, G, F = p
    ptot = logm0, G, F, .1
    return lnL.lnLikelihood(ptot)
```

```python
initial_binned = [14.11747161, -0.26151598,  0.72843539]
npath = 150
ndim = 3
nwalkers =70
pos_binned = initial_binned + 0.01 * np.random.randn(nwalkers, len(initial_binned))
```

```python
lnLikelihood(initial_binned)
```

```python
from multiprocessing import Pool
with Pool() as pool:
        sampler = emcee.EnsembleSampler(nwalkers, ndim, lnLikelihood, pool=pool)
        sampler.run_mcmc(pos_binned, npath,progress=True)
sampler_wl = sampler.get_chain(discard = 0, flat = True)
```
