```python
def lnLikelihood(p):
    log10m0, G, F, sigma_int = p
 
    if sigma_int < 0: return -np.inf

    if log10m0 < 0: return -np.inf
    #if sigma_int > 1: return -np.inf
    #sigma_int_ = np.sqrt(sigma_int**2 + F**2/match['obs_2'])
    #sigma_int = mr.sigma_lnM(match['obs_2'], match['redshift_2'], [aln, b, c])
    sigma_int_correct = np.sqrt(sigma_int**2 + F**2/match['obs_2'])
    for s in sigma_int_correct:
        if s < 0: return -np.inf
    mu_lnm = mr.lnM(match['obs_2'], match['redshift_2'], [log10m0, G, F]) - sigma_int_correct**2/2
    a= np.exp(-.5*(np.log(match['obs_1'])-mu_lnm)**2/(sigma_int_correct**2))/(match['obs_1']*np.sqrt(2*np.pi*sigma_int_correct))
    return np.sum(np.log(a))
```

```python
lnLikelihood([14,1,1, 1])
```

```python
nwalker=50
pos=.001*np.random.randn(nwalker,4) + np.array(np.array([14.1, 0, 1, 1,]))
sampler = emcee.EnsembleSampler(nwalker, 4, lnLikelihood,)
sampler.run_mcmc(pos, 200, progress=True);
flat_samples1 = sampler.get_chain(discard=10, thin=3, flat=True)
```
